# The Cantor-Lebesgue function

A continuous mapping from [0, 1] to [0, 1] that is non-differentiable on an uncountable set (the Cantor set).
<img src=https://upload.wikimedia.org/wikipedia/commons/7/7f/Cantor_function.gif width="500" height="500">

(GIF from Wikipedia)

```haskell
import Graphics.Rendering.Chart
import Data.Colour
import Data.Colour.Names
import Data.Default.Class
import Graphics.Rendering.Chart.Backend.Cairo
import Control.Lens
```

```haskell
cantor x x' y y' 0 = [(x, y), (x', y')]
cantor x x' y y' n = (cantor x x13 y y13 (n - 1)) ++ (cantor x23 x' y23 y' (n - 1))
                        where x13 = x + (x' - x) / 3
                              y13 = (y + y') / 2
                              x23 = x + 2 * (x' - x) / 3
                              y23 = (y + y') / 2
```

```haskell
values = plot_lines_values .~ [(cantor 0 1 0 1 10)]
              $ def
layout = layout_title .~ "The Cantor-Lebesgue function"
           $ layout_plots .~ [toPlot values]
           $ def
toRenderable layout
```
