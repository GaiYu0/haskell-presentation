# Haskell is pure!

Everything that cannot ensure referential transparency is impure.

Examples:

- IO

- Random number generator

- Concurrency (shared variable, channel, etc.)

```haskell
:t putStrLn
:t getLine
```

```haskell
import System.Random
```

```haskell
-- shared variable
import Control.Concurrent.MVar
:t newMVar
:t newEmptyMVar
:t takeMVar
:t putMVar
:t tryTakeMVar
:t tryPutMVar
```
